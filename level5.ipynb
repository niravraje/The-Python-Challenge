```python
import pickle
import urllib
```

```python
url = "http://www.pythonchallenge.com/pc/def/banner.p"
data_obj = urllib.request.urlopen(url)

print('Type of data_obj: ', type(data_obj))
```

pickle.load will de-serialize the data_object and return a python object!

```python
data = pickle.load(data_obj)
data
```

```python
for line in data:
    ans = [k*v for k,v in line]
    print(''.join(ans))
```
